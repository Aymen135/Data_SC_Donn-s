```python
n=int(input("donner la dimension du matrice : "))
matrix=[]

#Initialisation du matrice 

for i in range(n) :
    ligne=[]
    for j in range(n):
        ligne.append(int(input()))
    matrix.append(ligne)

# Affichage du matrice initial
for i in range(n):
    for j in range(n):
        print(matrix[i][j], end = " ")
    print()

```

```python
# 1er etape division de la ligne 1 par le coefficient A11
for i in range(1,n):
    matrix[0][i]=matrix[0][i]/matrix[0][0]

# Affichage du matrice Etape 1
for i in range(n):
    for j in range(n):
        print(matrix[i][j], end = " ")
    print()
```
